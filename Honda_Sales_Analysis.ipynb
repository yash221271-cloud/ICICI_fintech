```
import pandas as pd
df=pd.read_csv("honda_sales.csv")
df.head()
```

```
df.count()
```

```
df.info()
```

```
# region wise no. of orders placed
df.groupby('Region')["Order_ID"].count()
```

```
# monthwise total revenue
df.groupby('Month')["Sales_Amount"].sum()
```

```
# for each region car model wise total quantity sold
df.groupby(["Region","Car_Model"])["Quantity"].sum()
#sort values(ascending= False)
```

```
#display the records where quantity ordered > 15 and region is east
df.loc[(df["Quantity"]>=15) & (df["Region"]=="East")]
```

```
# plot the graph of model vs revenue
import matplotlib.pyplot as plt
plt.bar(df["Car_Model"],df["Sales_Amount"])
plt.xlabel("Car_Model")
plt.ylabel("Revenue")
plt.title("Model vs Revenue")
plt.show()
```

```
# plot the pie chart for region vs quantiity with labels
region_quantity = df.groupby("Region")["Quantity"].sum()
plt.pie(region_quantity, labels=region_quantity.index, autopct='%1.1f%%')
plt.title("Region vs Quantity")
plt.show()
```

```
# draw line chart for car model vs highest discount percentage
df.groupby("Car_Model")["Discount_Percent"].max().plot(kind="line")
plt.show()
```

```
# draw line chart for car model vs highest discount percentage
df.groupby("Car_Model")["Discount_Percent"].mean().plot(kind="line")
plt.show()
```
